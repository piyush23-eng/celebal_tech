```python
import pandas as pd

# Load the file
df = pd.read_csv("jackets.csv")
print(f"Dataset loaded successfully! Total rows: {df.shape[0]}")
# Safely fill the single missing discount value with 0
df['discount'] = df['discount'].fillna(0)

print("--- Missing Values Cleared ---")
print(df.isnull().sum())
# 1. Select specific columns
columns_to_keep = ['title', 'category', 'initial_price']
df_subset = df[columns_to_keep]
print("--- Selected Columns Subset ---")
display(df_subset.head())

# 2. Filter rows: Show jackets with an initial price greater than 2000
df_filtered = df[df['initial_price'] > 2000]
print(f"\nFiltered Dataset: Rows where initial_price > 2000 ({df_filtered.shape[0]} rows found)")
display(df_filtered.head())
# Drop any overlapping duplicate entries
df.drop_duplicates(inplace=True)

# Reset row numbers
df.reset_index(drop=True, inplace=True)
print(f"Duplicates handled. Total unique rows remaining: {df.shape[0]}")
import numpy as np

# Create simulated 'Quantity' column (between 1 and 3 items)
df['Quantity'] = np.random.randint(1, 4, size=len(df))

# Calculate total_amount = price * quantity
df['total_amount'] = df['initial_price'] * df['Quantity']

print("--- New 'total_amount' Column Created ---")
display(df[['title', 'initial_price', 'Quantity', 'total_amount']].head())
# Save out your new file
output_file = "cleaned_jackets_dataset.csv"
df.to_csv(output_file, index=False)

print(f"🎉 Success! Your final cleaned file is saved as: '{output_file}'")
```
